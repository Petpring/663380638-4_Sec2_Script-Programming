

## AI Ethics (จริยธรรมของ AI) คืออะไร

คือหลักการที่ทำให้แน่ใจว่า AI ที่เราสร้างขึ้นไม่สร้างความเสียหายต่อผู้คน เช่น:

- **ความเป็นธรรม (Fairness)** — โมเดลไม่ควรเลือกปฏิบัติกับคนกลุ่มใดกลุ่มหนึ่งโดยไม่เป็นธรรม เช่น ถ้าเราสร้างโมเดลคัดเลือกผู้สมัครงาน ต้องตรวจสอบว่าผลลัพธ์ไม่ลำเอียงทางเพศหรือเชื้อชาติ
- **ความเป็นส่วนตัว (Privacy)** — ข้อมูลที่ใช้ฝึกโมเดล (training data) ต้องได้รับความยินยอม และไม่เปิดเผยข้อมูลส่วนบุคคลโดยไม่จำเป็น
- **ความปลอดภัย (Safety)** — โมเดลต้องไม่ถูกนำไปใช้ในทางที่เป็นอันตราย

ตัวอย่างง่ายๆ ในเชิงโค้ด: เวลาเทรนโมเดลด้วย scikit-learn ก่อนใช้งานจริง ควรตรวจสอบ accuracy แยกตามกลุ่มย่อยของข้อมูล (subgroup) ไม่ใช่ดูแค่ accuracy รวม เพราะค่าเฉลี่ยรวมอาจซ่อนความลำเอียงของกลุ่มเล็กๆ ไว้

## Explainable AI (XAI) คืออะไร

คือการทำให้เราสามารถ "อธิบาย" ได้ว่าทำไมโมเดลถึงตัดสินใจแบบนั้น ไม่ใช่แค่บอกผลลัพธ์อย่างเดียว สำหรับมือใหม่ มีไลบรารีที่ใช้ง่ายสองตัวที่นิยมมากคือ `SHAP` และ `LIME`

ตัวอย่างโค้ดง่ายๆ การใช้ SHAP กับโมเดล RandomForest:

```python
import shap
from sklearn.ensemble import RandomForestClassifier

# สมมติว่าเทรนโมเดลแล้ว
model = RandomForestClassifier()
model.fit(X_train, y_train)

# สร้าง explainer และดูว่า feature ไหนมีผลต่อการทำนายมากที่สุด
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

shap.summary_plot(shap_values, X_test)
```

โค้ดนี้จะแสดงกราฟว่า "feature" ตัวไหน (เช่น อายุ, รายได้) มีผลต่อการทำนายของโมเดลมากหรือน้อยแค่ไหน ทำให้เราเข้าใจตรรกะเบื้องหลังโมเดล แทนที่จะเป็น black box

สำหรับโมเดลที่เรียบง่ายกว่า เช่น Logistic Regression หรือ Decision Tree การอ่านค่า coefficient หรือดูโครงสร้าง tree ตรงๆ ก็ถือเป็นรูปแบบหนึ่งของ explainability ที่ทำได้ง่ายกว่าและเหมาะกับมือใหม่

## Responsible AI คืออะไร

เป็นแนวทางปฏิบัติที่ครอบคลุมกว้างกว่า AI Ethics เล็กน้อย คือการนำหลักจริยธรรมไปใช้จริงตลอดวงจรชีวิตของโปรเจกต์ AI ตั้งแต่ขั้นตอนออกแบบจนถึงใช้งานจริง สิ่งที่มือใหม่ทำได้เลยมีดังนี้

หนึ่งคือเก็บ documentation ของข้อมูลที่ใช้ (data provenance) ว่าข้อมูลมาจากไหน มีข้อจำกัดอะไรบ้าง สองคือทดสอบโมเดลกับข้อมูลที่หลากหลาย ไม่ใช่แค่ข้อมูล test set เดียว สามคือเขียน comment และ docstring อธิบายว่าโมเดลใช้ทำอะไร มีข้อจำกัดอะไร เพื่อให้คนอื่นที่มาดูโค้ดต่อเข้าใจง่าย และสี่คือใช้ version control (เช่น git) บันทึกการเปลี่ยนแปลงของโมเดลและข้อมูล เพื่อตรวจสอบย้อนหลังได้ว่าทำไมโมเดลถึงให้ผลลัพธ์แบบนั้นในแต่ละช่วงเวลา

## คำแนะนำสำหรับมือใหม่

ถ้าคุณกำลังเริ่มเขียนโค้ด Python เกี่ยวกับ AI/ML ผมแนะนำให้เริ่มจากสามอย่างนี้ก่อน: ใช้ `pandas.DataFrame.describe()` และ `groupby()` ตรวจสอบข้อมูลก่อนเทรนเสมอว่ามีความลำเอียงไหม ใช้ไลบรารี `SHAP` หรือ `eli5` (ใช้ง่ายกว่า SHAP เล็กน้อย) เพื่อดูว่าโมเดลตัดสินใจอย่างไร และเขียน comment อธิบายทุกขั้นตอนสำคัญในโค้ด ไม่ใช่เพื่อให้คนอื่นเข้าใจอย่างเดียว แต่เพื่อให้ตัวเราเองย้อนมาตรวจสอบความถูกต้องได้ในอนาคต

ถ้าอยากให้ผมช่วยยกตัวอย่างโค้ดที่ใช้ข้อมูลจริง (เช่น dataset ที่คุณมี) เพื่อสาธิตการตรวจสอบ fairness หรือใช้ SHAP แบบเต็มรูปแบบ บอกผมได้เลยครับ

## 5. Code Snippet: "Learning Log" สำหรับติดตามการเรียนของตัวเอง (1 นาที)

เป็นตัวอย่างการใช้ `logging` module ของ Python เพื่อบันทึกความก้าวหน้าการเรียนของตัวเอง ฝึก habit การจด log ไว้ตั้งแต่สัปดาห์แรก — เป็นพื้นฐานที่ดีสำหรับ **Responsible AI/Coding practice** ในอนาคต (การมี record ว่าทำอะไรไปบ้าง ช่วยเรื่อง traceability และ accountability)

```python
import logging
from datetime import datetime

# ตั้งค่า logger เขียนลงไฟล์ learning_log.txt พร้อม timestamp
logging.basicConfig(
    filename="learning_log.txt",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    encoding="utf-8"
)

def log_progress(topic: str, status: str, note: str = ""):
    """
    บันทึกความก้าวหน้าการเรียนแต่ละหัวข้อ
    topic: ชื่อหัวข้อที่เรียน เช่น 'print() and variables'
    status: 'started' | 'practiced' | 'done'
    note: บันทึกเพิ่มเติม เช่น จุดที่ติดปัญหา
    """
    logging.info(f"[{topic}] status={status} note={note}")
    print(f"บันทึกแล้ว: {topic} -> {status}")

# ตัวอย่างการใช้งานหลังเรียนแต่ละหัวข้อใน Week 1
log_progress("print() and comments", "done")
log_progress("variables and data types", "practiced", note="ยังสับสน int vs str")
log_progress("input() and type conversion", "started")
```

**ทำไมต้องทำแบบนี้:** การมี log ไฟล์ช่วยให้ย้อนดูได้ว่าเรียนอะไรไปวันไหน ติดปัญหาอะไร ใช้เป็นข้อมูลตอนทำ portfolio หรือทบทวนก่อนสอบ Certificate (CISCO/Codecademy) ได้ง่ายขึ้น
