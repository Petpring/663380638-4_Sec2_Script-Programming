## 2. สิ่งที่เรียนในสัปดาห์นี้ (1 นาที)

| หัวข้อ | สิ่งที่ต้องจำ |
|---|---|
| Relational Operators | `==`, `!=`, `<`, `>`, `<=`, `>=` → ผลลัพธ์เป็น `bool` เสมอ |
| Logical Operators | `and` (ต้องจริงทั้งคู่), `or` (จริงอย่างน้อยหนึ่ง), `not` (กลับค่า) |
| `if-elif-else` | ตรวจตามลำดับ เจอ True ก่อนจะหยุดทันที เรียงเงื่อนไขให้ถูกสำคัญมาก |
| Nested Conditionals | if ซ้อน if ได้ แต่ระวังซ้อนลึกเกินจะอ่านยาก |
| List Comprehension | `[expr for x in iterable if condition]` แทน for-loop แบบสั้น เร็ว และอ่านง่ายกว่า (ถ้าไม่ซับซ้อนเกินไป) |

## 3. Code Snippets สำคัญ (2 นาที)

**Relational & Logical Operators**
```python
age = 25
has_license = True
eligible = (age > 21) and has_license   # True
print(f"Eligible for discount: {eligible}")
```

**if-elif-else: ระบบเกรด**
```python
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"
print(f"Grade: {grade}")   # Grade: B
```

**Nested if: ตัวอย่างจาก Lab 2.2 (แนะนำหนังตามอายุ)**
```python
age = 17
if age < 5:
    print("ยังเด็กเกินไป ดูการ์ตูนก่อนนะ")
elif age <= 12:
    print("แนะนำหนัง G หรือ PG")
elif age <= 17:
    print("แนะนำหนัง PG-13 (มีผู้ปกครองดูแล)")
else:
    print("ดูหนังเรทไหนก็ได้")
```

**List Comprehension: ย่อ for-loop ให้สั้นลง**
```python
# วิธีเดิม (for-loop)
squares = []
for x in range(10):
    squares.append(x ** 2)

# วิธีใหม่ (list comprehension) ทำงานเดียวกัน
squares = [x ** 2 for x in range(10)]
print(squares)   # [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# มี condition กรองค่าได้ด้วย
even_only = [x for x in range(20) if x % 2 == 0]
print(even_only)
```

## 5. Code Snippet: ต่อยอด "Learning Log" ของสัปดาห์ 2

ใช้ `logging` module เดิมจากสัปดาห์ 1 แต่เพิ่ม level การบันทึกตามความยาก เพื่อช่วยติดตามว่าหัวข้อไหนยังไม่แม่น (Control Flow ส่วนใหญ่มือใหม่จะพลาดเรื่อง logic ของ `elif` chain และ list comprehension ที่ซ้อนซับซ้อน)

```python
import logging

logging.basicConfig(
    filename="learning_log.txt",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    encoding="utf-8"
)

def log_progress(topic: str, status: str, difficulty: str = "ok", note: str = ""):
    """
    บันทึกความก้าวหน้าการเรียน พร้อมระบุระดับความยากที่รู้สึก
    difficulty: 'easy' | 'ok' | 'hard'  -> ใช้ทบทวนก่อนสอบว่าหัวข้อไหนต้องอ่านซ้ำ
    """
    level = logging.WARNING if difficulty == "hard" else logging.INFO
    logging.log(level, f"[{topic}] status={status} difficulty={difficulty} note={note}")
    print(f"บันทึกแล้ว: {topic} -> {status} (ความยาก: {difficulty})")

# ตัวอย่างการใช้งานหลังเรียน Week 2
log_progress("relational & logical operators", "done", difficulty="easy")
log_progress("if-elif-else chain ordering", "practiced", difficulty="ok",
             note="ลำดับเงื่อนไขสำคัญ ต้องเรียงจากเข้มงวดไปหลวม")
log_progress("list comprehension with condition", "started", difficulty="hard",
             note="งงเรื่อง syntax [expr for x in y if z]")
```

**ทำไมต้องเพิ่ม `difficulty`:** เวลาทบทวนก่อนสอบ Certificate (CISCO) สามารถกรองดูแต่เฉพาะ log ที่ `level=WARNING` เพื่อโฟกัสหัวข้อที่ยังไม่แม่นก่อน ไม่ต้องอ่านทุกอย่างซ้ำทั้งหมด — เป็นแนวทาง Responsible/Explainable practice แบบง่ายๆ คือ "รู้ตัวเองว่าทำไมถึงยังไม่เข้าใจ" ไม่ใช่แค่จดว่า "ทำแล้ว"
