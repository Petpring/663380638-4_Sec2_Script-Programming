```python
age = int(input("Please enter your age"))

if age < 5:
    sign = "You're too young for movies! Enjoy cartoons."

elif age <= 12:
    sign = "Recommended: G-rated or PG-rated movies."

elif age <= 17:
    sign = "Recommended: PG-13 or R-rated (with parental guidance)."

else:
    sign = "Recommended: Any movie rating."
print(sign)

like = input("Do you like action movie? (yes/no):")

if age >= 18 and like.lower() == "yes":
    print("You might enjoy the latest action blockbuster!")
```
