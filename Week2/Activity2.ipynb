```python
#Block A:Evaluation Check
temp = 25
is_raining = True

# ตรวจสอบว่าอุณหภูมิมากกว่า 20 และฝนไม่ตกหรือไม่

if temp > 20 and not is_raining:
# temp > 20 = True
# not is_raining = not True = False
# True and False = False
    print("Perfect day for a picnic!")

elif temp <= 20 or is_raining:
# temp <= 20 = False
# is_raining = True
# False or True = True
    print("Stay indoors or bring an umbrella.")

# ถ้าไม่เข้าเงื่อนไขใดเลยจึงจะมาที่ else
else:
    print("Unsure about the weather.")
```

```python
#Block B:Vectorized Filter
phrase = "Python 2026 Engine"

# ใช้ Set Comprehension
# - วนอ่านทีละตัวอักษร
# - เก็บเฉพาะตัวอักษรภาษาอังกฤษ (isalpha())
# - แปลงเป็นตัวพิมพ์เล็ก (lower())
# - ตัดสระ a, e, i, o, u ออก
# - ใช้ set เพื่อไม่ให้มีตัวอักษรซ้ำ
consonants = {
    c.lower()
    for c in phrase
    if c.isalpha() and c.lower() not in 'aeiou'
}

print(f"Unique consonants: {consonants}")

# ตัวอักษรที่ถูกเก็บคือ
# p, y, t, h, n, g
# (ตัว n พบหลายครั้ง แต่ set จะเก็บเพียงครั้งเดียว)

# OUTPUT ({'g', 'n', 'h', 't', 'y', 'p'})
```

```python
#  Choose Your Own Adventure : เลือกซื้อเบเกอรี่

# พิมพ์ข้อความต้อนรับผู้ใช้
print("ยินดีต้อนรับสู่ร้านเบเกอรี่ 🍰")
# แสดงรายการเมนูที่มีให้เลือก
print("วันนี้คุณอยากกินอะไร?")
print("1. เค้ก")
print("2. ขนมปัง")
print("3. ทาร์ต")

# รับอินพุตจากผู้ใช้สำหรับตัวเลือกเมนู
choice = input("เลือกเมนูไหนดี ? ")

# ตรวจสอบตัวเลือกของผู้เล่นโดยใช้ 'or' สำหรับเงื่อนไขหลายตัว
if choice == "เค้ก" or choice == "1":
    # ถ้าผู้ใช้เลือกเค้ก
    print("คุณเลือกเค้ก 🎂")

    # รับอินพุตสำหรับรสชาติเค้ก
    flavor = input("เลือกหน้ารสชาติ: 1.ช็อกโกแลต หรือ 2.สตรอว์เบอร์รี : ")

    # ตรวจสอบรสชาติที่เลือก
    if flavor == "ช็อกโกแลต" or flavor == "1":
        print("คุณได้เค้กช็อกโกแลต อร่อยมาก! 🍫")
    elif flavor == "สตรอว์เบอร์รี" or flavor == "2" :
        print("คุณได้เค้กสตรอว์เบอร์รี หวานอมเปรี้ยว 🍓")
    else:
        print("ไม่มีรสชาตินี้")

elif choice == "ขนมปัง" or choice == "2":
    # ถ้าผู้ใช้เลือกขนมปัง
    print("คุณเลือกขนมปัง 🍞")

    # รับอินพุตสำหรับไส้ขนมปัง
    bread = input("เลือกไส้: 1.ไส้กรอก หรือ 2.ครีม : ")

    # ตรวจสอบไส้ที่เลือก
    if bread == "ไส้กรอก"or bread == "1":
        print("คุณได้ขนมปังไส้กรอก อิ่มอร่อย 🌭")
    elif bread == "ครีม" or bread == "2" :
        print("คุณได้ขนมปังครีม หอมหวาน 🥛")
    else:
        print("ไม่มีไส้นี้")

elif choice == "ทาร์ต" or choice == "3":
    # ถ้าผู้ใช้เลือกทาร์ต
    print("คุณเลือกทาร์ต 🥧")

    # รับอินพุตสำหรับประเภททาร์ต
    tart = input("เลือก: 1.ไข่ หรือ 2.ผลไม้ : ")

    # ตรวจสอบประเภทที่เลือก
    if tart == "ไข่" or tart == "1":
        print("คุณได้ทาร์ตไข่ หอมมัน 🥚")
    elif tart == "ผลไม้" or tart == "2":
        print("คุณได้ทาร์ตผลไม้ สดชื่น 🍓")
    else:
        print("ไม่มีเมนูนี้")

else:
    # ถ้าผู้ใช้ป้อนอย่างอื่น
    print("วันนี้คุณตัดสินใจไม่กิน แล้วเดินกลับบ้าน")
```
